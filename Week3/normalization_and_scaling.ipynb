```python
import pandas as pd
from sklearn.preprocessing import StandardScaler,MinMaxScaler

iris_data = pd.read_csv('Iris.csv')
```

```python
scaler = StandardScaler()
iris_standarized = scaler.fit_transform(iris_data.iloc[:,:-1])

min_max_scaler = MinMaxScaler()
iris_min_max_scaled = min_max_scaler.fit_transform(iris_data.iloc[:,:-1])

```

```python
#visualization 
# histogram and density plots
import seaborn as sns
import matplotlib.pyplot as plt
        
feature = 'SepalLengthCm'
plt.figure(figsize=(20,5))
plt.subplot(1, 3, 1)
sns.histplot(iris_data[feature], kde=True, color='blue')
plt.tight_layout()
plt.title(f'Original {feature}')



plt.subplot(1, 3, 2)
sns.histplot(iris_standarized[:, 0], kde=True, color='green')  # Adjust index accordingly
plt.title(f'Standardized {feature}')
plt.tight_layout()

plt.subplot(1,3,3)
sns.histplot(iris_min_max_scaled,kde=True,color='red')
plt.title(f'Scaled {feature}')
plt.tight_layout()

plt.show()
```

```python
#Box plot
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
sns.boxplot(data=iris_data,y=feature)
plt.title(f'Original {feature}')


plt.subplot(1,3,2)
sns.boxplot(y=iris_standarized[:,0])
plt.title(f'Standarized {feature}')



plt.subplot(1,3,3)
sns.boxplot(y=iris_min_max_scaled)
plt.title(f'Scaled {feature}')
plt.tight_layout()
plt.show()
```

```python
#scatter plot
# A scatter plot helps to determine whether there is a relationship or correlation between two variables.
# It helps identify linear or non-linear relationships
plt.figure(figsize=(12,4))
plt.scatter(iris_data['SepalLengthCm'], iris_data['SepalWidthCm'], alpha=0.5, label='Original')
plt.scatter(iris_min_max_scaled[:,0], iris_min_max_scaled[:,0], alpha=0.5, label='Scaled')
plt.scatter(iris_standarized[:, 0], iris_standarized[:, 1], alpha=0.5, label='Standardized')  # Adjust indices accordingly

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs. Width: Original vs. Standardized vs Scaled')
plt.legend()
plt.show() 
```
